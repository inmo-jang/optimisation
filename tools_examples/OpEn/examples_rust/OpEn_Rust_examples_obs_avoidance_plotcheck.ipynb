```python
import matplotlib.pyplot as plt
import numpy as np

def plot(p, centre):
    fig = plt.figure()
    ax = fig.add_subplot(111)

    # Cost function
    x = np.arange(-5, 5, 0.0001) 
    y = p[0]*(x - p[1])**2 + p[2]    
    cost, = ax.plot(x,y)
        
    # Add Circle
    radius = 1
    circle1 = plt.Circle((centre[0], centre[1]), radius, color='r')
    ax.add_artist(circle1)

    # Axis Control
    ax.axis('equal')
    ax.set(xlim=(-3, 3), ylim=(-3, 3))
    ax.grid(color='k', linestyle='-.', linewidth=0.5)
    plt.show()

```

```python
plot([0.5, 0.0, 0.0],[0.0, 0.0])
```

```python
plot([0.5, 0.0, 0.0],[-0.25, 0.0])
```
